```python
# Helper function
def titlelize(word):
    sect = word.split(' ')
    return ' '.join([i.title() for i in sect])
```

```python
# Get male names and distribution
df_male = pd.read_csv('male_names.csv').dropna(inplace=False)
tot_freq = 0
for index, row in df_male.iterrows():
    tot_freq += row['frequency']

male_probs = df_male['frequency'].apply(lambda x: x/tot_freq).tolist()
male_names = df_male['name'].apply(lambda x: titlelize(x)).tolist()
male_names, male_probs
```

```python
# Get female names and distribution
df_fem = pd.read_csv('female_names.csv').dropna(inplace=False)
tot_freq = 0
for index, row in df_fem.iterrows():
    tot_freq += row['frequency']

fem_probs = df_fem['frequency'].apply(lambda x: x/tot_freq).tolist()
fem_names = df_fem['name'].apply(lambda x: titlelize(x)).tolist()
fem_names, fem_probs
```

```python
# Get surnames and distribution
df_sur = pd.read_csv('surnames_freq_ge_100.csv').dropna(inplace=False)
tot_freq = 0
for index, row in df_sur.iterrows():
    tot_freq += row['frequency_first']

sur_probs = df_sur['frequency_first'].apply(lambda x: x/tot_freq).tolist()
surnames = df_sur['surname'].apply(lambda x: titlelize(x)).tolist()
surnames, sur_probs
```
